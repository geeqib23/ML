```python
import pandas as pd
import numpy as np

# UnicodeDecodeError: 'utf-8' codec can't decode byte 0xa0 in position 0: invalid start byte
df = pd.read_csv("datasets/Minimum Wage Data.csv", encoding="latin")
df.to_csv("datasets/minwage.csv", encoding="utf-8", index = False)
df.set_index("Year").head()
```

```python
gb = df.groupby("State")
gb.get_group("Alabama").set_index("Year").head()
```

```python
act_min_wage = pd.DataFrame()
# Iterating over groups
for name, group in df.groupby("State"):
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year")[["State.Minimum.Wage.2020.Dollars"]].rename(columns={"State.Minimum.Wage.2020.Dollars":name})
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year")[["State.Minimum.Wage.2020.Dollars"]].rename(columns={"State.Minimum.Wage.2020.Dollars":name}))

# act_min_wage.head()
# OR
# pd.pivot_table(df, values="State.Minimum.Wage.2020.Dollars", index="Year", columns='State')
# OR
df.groupby(["Year","State"]).last().unstack()
```
